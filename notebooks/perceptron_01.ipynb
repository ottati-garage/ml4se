必要なモジュールをインポート

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import Series, DataFrame
````

トレーニング用データのインポート

* 文字が0 -> -1
* 文字が1 -> 1

````python
def import_data(filename_images, filename_labels):
    train_set = pd.read_csv(filename_images, header=None)
    labels_data = pd.read_csv(filename_labels, header=None)
    train_set['type'] = labels_data[0]
    train_set['type'] = train_set['type'].replace(0, -1)
    return train_set
````

````python
train_set = import_data("sample-images-01.txt", "sample-labels-01.txt")
print "len(train_set):", len(train_set)
train_set[118:126]
````

訓練データからバイアスを求める関数

````python
def calc_bias(train_set):
    return (1.0 / 784) * sum([train_set[i].mean() for i in range(784)])

# test
calc_bias(train_set)
````

以下を計算する関数

```
w0 * bias + w1 * x0 + w2 * x1 + ... + wn * xn-1
```

````python
def calc_partial_E(w0, W, bias, char):
    # w0
    total = w0 * bias
    # w1 ~ wn
    total += (char[0:784] * W).sum()
    r = char['type'] * total
    return r

# test
w0 = 3
W = [1] * 784
bias = calc_bias(train_set)
for i, char in train_set[3:6].iterrows():
    print calc_partial_E(w0, W, bias, char)
````

学習用関数

````python
def run_train(train_set):
    w0 = 0
    W = [0] * 784
    bias = calc_bias(train_set)
    paramhist = DataFrame([W])
    for i in range(10):
        for j, char in train_set.iterrows():
            r = calc_partial_E(w0, W, bias, char)
            if r <= 0:
                # update w0 and W
                w0 += char['type'] * 1
                for k in range(len(W)):
                    W[k] += char['type'] * char[k]
        paramhist = paramhist.append(Series([w0] + W), ignore_index=True)
    
    # 判定誤差の計算
    err = 0
    for index, char in train_set.iterrows():
        if calc_partial_E(w0, W, bias, char) <= 0:
            err += 1
    err_rate = err * 100.0 / len(train_set)
    return paramhist, err_rate
````

学習させる

````python
paramhist, err_rate = run_train(train_set)
print err_rate
````

### 他の文字も読み込ませてみる

0-9の数字で判定した割合をそれぞれの数値で出す

````python
spam_set = import_data("sample-images.txt", "sample-labels.txt")
spam_set = spam_set[:10000]
print "len(spam_set):", len(spam_set)
spam_set['type'] = spam_set['type'].replace(-1, 0)  # -1になっている0を-1に戻す
spam_set[138:146]
````

その数字を0と判定したか1と判定したかを％で算出

````python
r0 = [0] * 10
r1 = [0] * 10

dic1 = dict(zip(range(10), [0] * 10))
bias = calc_bias(spam_set)
last_param = list(paramhist[-1:].values[0])
w0 = last_param[0]
W = last_param[1:]
for index, char in spam_set.iterrows():
    # w0
    total = w0 * bias
    # w1 ~ wn
    total += (char[0:784] * W).sum()
    if total <= 0:
        r0[char['type']] += 1
    else:
        r1[char['type']] += 1
        
r = []
for i in range(10):
    total = r0[i] + r1[i]
    p_0 = round((r0[i] / float(total)) * 100, 1)
    p_1 = round((r1[i] / float(total)) * 100, 1)
    r.append([p_0, p_1])
r_df = DataFrame(r, columns=["0と判定 (%)", "1と判定 (%)"])
r_df
````
